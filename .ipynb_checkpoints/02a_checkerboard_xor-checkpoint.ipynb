### Representational Power, Logic Gates


Neural networks have the capacity to model very complex functions.
In fact, they have nearly unlimited representational power.
However, realizing this capacity for a given problem is limited by finite data and limitations in the gradient descent learning algorithms.  Only pieces of this are understood, and this is very much an area of active research.
(Side note: in searching through the literature, be sure to distinguish between 'reprentation' theorems (exact) and 'approximation' theorems, and, moreoever, note that these theorems do not provide an algorithm for realizing the representations or approximations that they prove to exist.)

Here, let's consider the representational power to model logic functions.    

<table>
<tr>
<th>Gate</th>
<th>Formula</th>
</tr>
<tr>
<td>AND</td>
<td>$\sigma( M (X_1+\dots+X_k)-k M ) \geq 0.5$ </td>
</tr>
<tr>
<td>OR</td>
<td>$\sigma( M (X_1+\dots+X_k) - k M ) \geq 0.5$ </td>
</tr>
<tr>
<td>NOT</td>
<td>$\sigma( M - 2M X ) \geq 0.5$ </td>
</tr>
<tr>
<td>NAND</td>
<td>$\sigma( k M - M (X_1+\dots+X_k) ) \geq 0.5$ </td>
</tr>
<tr>
<td>.</td>
<td> ------------------------------------------------------------------------  </td>
</tr>
</table>

where $\sigma$ is the sigmoid/logistic function (or a "generalized sigmoid function")
and $M \gg 1$ is a very large number.  Notice that these are the general logic gates with multiple inputs and include binary logic gates as a special case.

*Most importantly, NAND is included here.  NAND is universal for logic functions in the sense that all logic functions can be written as a composition of NAND gates.  Therefore, **all logic functions can be reprensented functionally as a feedforward network with the above NAND unit.** *


In the example below, we see that a net easily learns the XOR gate from data.

```python
import numpy as np
#
%matplotlib inline
import matplotlib
matplotlib.rc('font', **{'family':'Times New Roman', 'weight':'normal', 'size':10})
import matplotlib.pyplot as plt
```

```python
ng = 2   # the number of checks along a dimension
t = np.linspace(0,1,ng+1)
```

```python
X = np.random.uniform(0,1,size=(1000,2))
```

```python
Y = np.sum(np.floor(X*ng),axis=1) % 2
Y = np.vstack([Y, 1.-Y]).T   # have 2-dimensions for cross-entropy in neural net
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(4,4)
for s in t:
    plt.plot(s*np.ones_like(t),t,color='black',linestyle='--',linewidth=1.)
    plt.plot(t,s*np.ones_like(t),color='black',linestyle='-',linewidth=1.)
plt.scatter(X[:,0],X[:,1],s=50,c=Y[:,0],edgecolors='',cmap=plt.cm.coolwarm)
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
import tensorflow as tf
```

```python
def fc_layer(incoming, size_in, size_out, name="fc"):
    with tf.name_scope(name):
        w = tf.Variable(tf.truncated_normal([size_in, size_out], stddev=0.1), name="W")
        b = tf.Variable(tf.constant(0.1, shape=[size_out]), name="B")
        act = tf.nn.relu(tf.matmul(incoming, w) + b)
        tf.summary.histogram("weights", w)
        tf.summary.histogram("biases", b)
        tf.summary.histogram("activations", act)
        return act
```

```python
tf.reset_default_graph()
sess = tf.Session()
```

```python
# Setup placeholders, and reshape the data
x = tf.placeholder(tf.float32, shape=[None, 2], name="x")
y = tf.placeholder(tf.float32, shape=[None, 2], name="labels")
```

```python
fc1 = fc_layer(x, 2, 16, 'fc1')
fc2 = fc_layer(fc1, 16, 16, 'fc2')
logits = fc_layer(fc2, 16, 2, 'fc3')
```

```python
with tf.name_scope("xent"):
    xent = tf.reduce_mean(
        tf.nn.softmax_cross_entropy_with_logits(
            logits=logits, labels=y), name="xent")
    tf.summary.scalar("xent", xent)
    
with tf.name_scope("train"):
    learning_rate = 1.e-3
    train_step = tf.train.AdamOptimizer(learning_rate).minimize(xent)

with tf.name_scope("accuracy"):
    correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(y, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    tf.summary.scalar("accuracy", accuracy)

summ = tf.summary.merge_all()    
```

```python
saver = tf.train.Saver()
```

```python
sess.run(tf.global_variables_initializer())
writer = tf.summary.FileWriter('/tmp/checkerboard/1')
writer.add_graph(sess.graph)
```

```python
import os
```

```python
num_steps = 2000
batch_size = 128
for step in range(num_steps):
    offset = (step * batch_size) % X.shape[0]
    # Generate a minibatch.
    batch_data = X[offset:(offset + batch_size), :]
    batch_labels = Y[offset:(offset + batch_size), :]
    if step % 5 == 0:
        [train_accuracy, s] = sess.run([accuracy, summ], feed_dict={x: batch_data, y: batch_labels})
        writer.add_summary(s, step)        
    if step % 500 == 0:
        print("Minibatch accuracy: %.2f" % train_accuracy)
        saver.save(sess, os.path.join('/tmp/checkerboard', "model.ckpt"), step)
    sess.run(train_step, feed_dict={x: batch_data, y: batch_labels})
```

```python
decision_values, predicted = sess.run([logits, tf.nn.softmax(logits)], feed_dict={x: X})
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(4,4)
for s in t:
    plt.plot(s*np.ones_like(t),t,color='black',linestyle='--',linewidth=1.)
    plt.plot(t,s*np.ones_like(t),color='black',linestyle='-',linewidth=1.)
plt.scatter(X[:,0],X[:,1],s=50,c=predicted[:,0],edgecolors='',cmap=plt.cm.coolwarm)
plt.axis('equal')
plt.axis('off')
plt.show()
```
